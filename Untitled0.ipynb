# 1: Loading the Dataset

a: Load the datasets


---


We will load the PlayerStatistics.csv and Games.csv datasets into pandas dataframes so that we can analyze player and game data.
A warning may appear regarding column data types, which does not affect our analysis.



```python
import pandas as pd

# Load the datasets
player_stats = pd.read_csv('PlayerStatistics.csv', low_memory=False)
games = pd.read_csv('Games.csv', low_memory=False)

print("Datasets loaded successfully!")
```

b: Display first 10 rows
Display the first 10 rows of each dataset to understand their structure and data.

```python
print("Player Statistics Dataset (first 10 rows):")
display(player_stats.head(10))


print("Games Dataset (first 10 rows):")
display(games.head(10))
```

# 2: Data Manipulation

a: Check for missing values
Check for any missing values in both datasets.

```python
# Check for missing values in PlayerStatistics.csv
missing_player_stats = player_stats.isnull().sum()
print("Missing values in PlayerStatistics.csv:")
display(missing_player_stats)

# Check for missing values in Games.csv
missing_games = games.isnull().sum()
print("\nMissing values in Games.csv:")
display(missing_games)
```

b: Create Full Name column


---
Combine firstName and lastName to create a Full Name column.


```python
# Combine firstName and lastName into a new column 'Full Name'
player_stats['Full Name'] = player_stats['firstName'] + ' ' + player_stats['lastName']

# Display first 5 rows to confirm the new column
print("First 5 rows with Full Name column:")
display(player_stats[['firstName', 'lastName', 'Full Name']].head())
```

c: Update gameType field


---

Update the gameType field to replace 'Preseason' with 'Pre-season'.

```python
# Replace 'Preseason' with 'Pre-season' in the gameType column
player_stats['gameType'] = player_stats['gameType'].replace('Preseason', 'Pre-season')

# Display unique values of gameType to confirm the change
print("Unique game types after update:")
display(player_stats['gameType'].unique())
```

# 4: Players Analysis

a: Top 10 players based on total points


---


Create a bar chart showing top 10 players based on total points from 1947-2025.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress future warnings for Seaborn
warnings.filterwarnings("ignore", category=FutureWarning)

# Top 10 players by total points
top10_points = player_stats.groupby('Full Name')['points'].sum().sort_values(ascending=False).head(10)
top10_df = top10_points.reset_index()
top10_df.columns = ['Player', 'Points']

# Bar chart
sns.barplot(data=top10_df, x='Points', y='Player', palette="viridis")
plt.xlabel('Total Points')
plt.ylabel('Player')
plt.title('Top 10 Players by Total Points (1947-2025)')
plt.show()
```

b: Top 5 players in 2025 by steals


---


Visualize top 5 players in 2025 based on steals.


```python
# Convert gameDate to datetime
player_stats['gameDate'] = pd.to_datetime(player_stats['gameDate'], errors='coerce')

# Filter 2025 games
player_2025 = player_stats[player_stats['gameDate'].dt.year == 2025]

# Get top 5 players by steals
top5_steals = (
    player_2025.groupby('Full Name')['steals']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# Plot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(data=top5_steals, x='steals', y='Full Name', palette='magma')
plt.title('Top 5 Players by Steals in 2025')
plt.xlabel('Total Steals')
plt.ylabel('Player')
plt.show()
```

c: Total points per year (2015-2025)


---


Line chart displaying total points scored each year from 2015-2025.


```python
# Make sure gameDate is datetime
player_stats['gameDate'] = pd.to_datetime(player_stats['gameDate'], errors='coerce')

# Extract year
player_stats['Year'] = player_stats['gameDate'].dt.year

# Filter between 2015 and 2025
points_per_year = (
    player_stats[(player_stats['Year'] >= 2015) & (player_stats['Year'] <= 2025)]
    .groupby('Year')['points']
    .sum()
)

# Line chart
plt.figure(figsize=(8,5))
plt.plot(points_per_year.index, points_per_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Points')
plt.title('Total Points per Year (2015-2025)')
plt.grid(True)
plt.show()
```

Display total points scored by Shai Gilgeous-Alexander in 2025.

# 5: Games Analysis

a: Top 3 games by attendance


---


Display top 3 games by attendance including home and away team names.


```python
# Ensure attendance is numeric
games['attendance'] = pd.to_numeric(games['attendance'], errors='coerce')

# Sort by attendance descending and pick top 3
top3_attendance = games.sort_values(by='attendance', ascending=False).head(3)

# Select only relevant columns
top3_attendance_df = top3_attendance[['hometeamName', 'awayteamName', 'attendance']]

# Display the dataframe
top3_attendance_df
```

b: Pie chart of top 5 teams winning percentage in 2025


---


Pie chart showing winning percentage of top 5 teams in 2025; others grouped as 'Others'.


```python
# Ensure gameDate is datetime
games['gameDate'] = pd.to_datetime(games['gameDate'], errors='coerce')

# Extract year
games['year'] = games['gameDate'].dt.year

# Filter for 2025 games
games_2025 = games[games['year'] == 2025]

# Count wins per team
wins_count = games_2025['winner'].value_counts()

# Top 5 teams
top5 = wins_count.head(5)

# Sum the rest as "Others"
others = wins_count[5:].sum()
top5['Others'] = others

# Pie chart
plt.figure(figsize=(7,7))
plt.pie(
    top5,
    labels=top5.index,
    autopct=lambda p: f'{p:.1f}%\n({int(round(p*sum(top5)/100))})',
    startangle=140,
    colors=sns.color_palette("pastel")
)
plt.title('Winning Percentage of Top 5 Teams in 2025 (Others combined)')
plt.show()
```
